```python
for dado in dados[0:11]:
    ded = np.loadtxt(dado, comments = '%')
    dados_anos= ded[:,0]+1/12*(ded[:,1]-1)
    anual = ded[0:,4]
    plt.plot(dados_anos,anual,'r-')
    plt.xlabel('Anos')
    plt.ylabel('Mdia de temperatura anual(C) ')
    plt.title(dado[6:19])
    plt.show()
    
# Importamos todos os dados e atribumos a ded
#dados_anos= ded[:,0]+1/12*(ded[:,1]-1)
#importamos os dados de ano e ja fizemos o calculo para nao atribuir varios valores de anomalias ao mesmo ano
#anual = ded[0:,4]
#impotamos os dados de anomalia anual
#plt.plot(dados_anos,anual,'r-')
#plt.xlabel('Anos')
#plt.ylabel('Mdia de temperatura anual(C) ')
#plt.title(dado[6:19]) definimos as strings para cada um aparecer o nome da estao de plotagem
# Definimos o grafico da funo e demos o titulo e as variaveis do grafico.
    
    
```
